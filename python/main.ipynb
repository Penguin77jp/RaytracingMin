```python
import cppmodule
import numpy as np
import matplotlib.pyplot as mplot
import pandas
import plotly.express as px
import plotly.graph_objects as go
```

```python
settingDataPath = "../camera/convex.json"

resultData = cppmodule.cameraLensSample(settingDataPath, [1,0,0], 1000)

plot = []
for line in resultData :
    xData = []
    yData = []
    zData = []
    for point in line :
        xData.append(point[0])
        yData.append(point[1])
        zData.append(point[2])
    if len(xData) == 0:
        continue
    plot.append(go.Scatter3d(x=xData, y=yData, z=zData, mode='lines'))

fig = go.Figure(data=plot)
fig.show()
```

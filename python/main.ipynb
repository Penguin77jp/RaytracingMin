```python
import cppmodule
import numpy as np
import plotly.io as pio
pio.renderers.default = 'notebook_connected'
import matplotlib.pyplot as mplot
import pandas
import plotly.express as px
import plotly.graph_objects as go
import numpy
```

```python
settingDataPath = "../settingData.json"
lensSystemPath = "../camera/convex.json"

resultData = cppmodule.cameraLensSample(settingDataPath, lensSystemPath, [4,0,0], [-0.15, 0, 1], -30)

plot = []
for line in resultData :
    xData = []
    yData = []
    zData = []
    print(line)
    for point in line :
        print(point)
        xData.append(point[0])
        yData.append(point[1])
        zData.append(point[2])
    if len(xData) == 0:
        continue
    plot.append(go.Scatter3d(x=xData, y=yData, z=zData, mode='lines'))

fig = go.Figure(data=plot)
fig.show()
```

```python
# check focal distance

import random as rand


def random_n1_p1() :
    return 2.0 * rand.random() - 1.0


settingDataPath = "../settingData.json"
lensSystemPath = "../camera/convex.json"

plot = []
for i in range(50) :
    samplePoint = [random_n1_p1(), random_n1_p1(), random_n1_p1()]
    resultData = cppmodule.cameraLensSampleRayDir(settingDataPath, lensSystemPath, samplePoint, 1)

    for line in resultData :
        xData = []
        yData = []
        zData = []
        for point in line :
            xData.append(point[0])
            yData.append(point[1])
            zData.append(point[2])
        if len(xData) == 0:
            continue
        plot.append(go.Scatter3d(x=xData, y=yData, z=zData, mode='lines'))

fig = go.Figure(data=plot)
fig.show()
```
